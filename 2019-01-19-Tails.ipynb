## 1.1 thin tailed distribution
Probability distribution functions that decay faster than an exponential are called thin-tailed distributions.
Example: normal distribution

## 1.2 heavy tailed distribution
Probability distribution functions that decay slower than an exponential are called heavy-tailed distributions.
Example: t distribution

### 1.2.1 fat-tailed distribution
fat-tailed distributions is that the mean, variance, and other measures that describe the shape of the distribution might not be defined  
Example: Pareto distribution / Cauchy distribution

```python
import numpy as np
a, m = 3., 2.  # shape and mode
s = (np.random.pareto(a, 1000) + 1) * m
```

```python
plt.plot(np.arange(len(s)),s)
```

```python
import matplotlib.pyplot as plt
count, bins, _ = plt.hist(s, 100, density=True)
fit = a*m**a / bins**(a+1)
plt.plot(bins, max(count)*fit/max(fit), linewidth=2, color='r')
```
